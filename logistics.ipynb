```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
fig = plt.figure(figsize=(100, 100))
excreta = np.linspace(1,100,100)
enzyme = np.linspace(0.01,1,100)

def f(a,b):
    return (0.02036*a + 5.085)/b
t = []
for i in range(100):
    t.append(f(excreta[i],enzyme[i]))


fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot3D(excreta,enzyme,t, c='r', marker='o')
ax.set_xlabel("excreta in kg",fontsize = 10)
ax.set_ylabel("enzyme in uM",fontsize = 10)

ax.set_zlabel("time in minutes",fontsize = 10)
plt.colorbar();
plt.show()
```
